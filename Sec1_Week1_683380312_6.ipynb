2. Declare and print variables:


```python
Age = 67
GPA = 3.67
Favorite_Color = "Black"
Student = True

print("Age = ",Age ,type(Age))
print("GPA = ",GPA,type(GPA))
print("Favorite_Color = ",Favorite_Color,type(Favorite_Color))
print("Student = ",Student,type(Student))
```

3. Perform simple arithmetic:


```python
num1 = 20
num2 = 20
num3 = 20
num4 = 7
sum = num1 + num2 + num3 + num4
diff = num1 - num4
Prod = num1 * num2
Div = num1/num2

print("Sum = ",sum)
print("difference = ",diff)
print("Product = ",Prod)
print("Division = ",int(Div))
```

4. String concatenation and f-strings:


```python
First_name = "BAMBAM"
Last_name = "WILASINEE"
Full_name = First_name + " " + Last_name
print(f"Hello,{Full_name}")
```
